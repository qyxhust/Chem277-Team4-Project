```python
import pandas as pd
```

```python
ad_ = pd.read_csv('data/raw/ad_data.csv')
pd_ = pd.read_csv('data/raw/pd_data.csv')
ad_ = ad_.dropna(axis=1, how='all')
pd_ = pd_.dropna(axis=1, how='all')

print(ad_.shape)
print(pd_.shape)

# contact two tables
test_data = pd.merge(ad_, pd_, on='Protein', suffixes=('_ad', '_pd'))
test_data.head()

test_data.to_csv('data/raw/raw_data.csv', index=False)

```

```python
test_data.head()
```

```python
test_data = test_data[['Protein', 'HR[95%CI]_ad', 'P_value_ad','HR[95%CI]_pd', 'P_value_pd']]

import numpy as np
# 以P_value_ad为例，取对数后覆盖原列
test_data['PD_logp'] = test_data['P_value_ad'].apply(np.log)
test_data['AD_logp'] = test_data['P_value_pd'].apply(np.log)

# 将HR列按照[分开，取第一块
test_data['ad_value'] = test_data['HR[95%CI]_ad'].str.split('[').str[0].str.strip()
test_data['pd_value'] = test_data['HR[95%CI]_pd'].str.split('[').str[0].str.strip()

test_data['AD_beta'] = test_data['ad_value'].astype(float).apply(np.log)
test_data['PD_beta'] = test_data['pd_value'].astype(float).apply(np.log)


filered_data = test_data[['Protein', 'AD_beta', 'AD_logp', 'PD_beta', 'PD_logp']]

print(filered_data.head())

filered_data.to_csv('data/raw/filtered_data.csv', index=False)



```
