```python
def eval_model_detailed(model, data, mask):
    model.eval()
    with torch.no_grad():
        pred_ad, pred_pd, pred_ftd, pred_als, _ = model(data)

        y = data.y

        loss_ad  = F.mse_loss(pred_ad[mask],  y[mask, 0].unsqueeze(1))
        loss_pd  = F.mse_loss(pred_pd[mask],  y[mask, 1].unsqueeze(1))
        loss_ftd = F.mse_loss(pred_ftd[mask], y[mask, 2].unsqueeze(1))
        loss_als = F.mse_loss(pred_als[mask], y[mask, 3].unsqueeze(1))

    return {
        "total": (loss_ad + loss_pd + loss_ftd + loss_als).item(),
        "AD": loss_ad.item(),
        "PD": loss_pd.item(),
        "FTD": loss_ftd.item(),
        "ALS": loss_als.item(),
    }
```
